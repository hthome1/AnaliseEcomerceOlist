### Nesse documento será estudado um dataset feito pela Olist Store, que contém informações acerca de 100 mil compras onlines feitas entre 2016 e 2018 em multiplos marketplaces no Brasil. A pergunta motriz da pesquisa será: Quais são os fatores acionáveis pela empresa que tipicamente levam clientes a darem avaliações ruins em compras? 

Bibliotescas utilizadas no estudo de caso

````python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.corpus import stopwords
import pandas as pd
````

### Fazendo analise exploratoria dos reviews

Primeiro passo será binarizar a nota do review, usando o limiar de 3. Assim, temos que se a nota for menor ou igual a 3 é negativa e portanto tem o valor 0. Caso contrário ela é positiva e recebe o valor 1 

````python
df = pd.read_csv("olist_order_reviews_dataset.csv")
````

````python
df = df[["review_score", "review_comment_title", "review_comment_message"]]
````

````python
df["review_score"].loc[df["review_score"] <= 3] = 0
````

````python
df["review_score"].loc[df["review_score"] > 3] = 1
````

Juntando o titulo com o corpo do review, uma vez que ambos contém informaçoes pertinentes e nem sempre ambos são preenchidos

````python
df["review"] = df["review_comment_title"].fillna("& ") + df["review_comment_message"].fillna("& ")
````

````python
df = df[df.review != "& & "]
````

````python
positivo = len(df["review_score"].loc[df["review_score"] == 1])
negativo = len(df["review_score"].loc[df["review_score"] == 0])
````

````python
tipo = ["positivo", "negativo"]
quantidade = [positivo,negativo]
plt.pie(quantidade,labels = tipo, autopct = '%1.1f%%')
plt.legend()
plt.show()
````

### Quase 2/3 das avaliações são positivas e 1/3 negativas

````python
# nltk.download('stopwords')
sw = stopwords.words('portuguese')
sw.append("&")
````

````python
# PLOTANDO GRAFICO 1
avaliacões = [av for av in df["review"]]
vectorizer = CountVectorizer(stop_words = sw)
x = vectorizer.fit_transform(avaliacões)
x_ = np.sum(x, axis=0)

tuplas = [ (x_[0,vectorizer.vocabulary_[i]], i) for i in vectorizer.vocabulary_.keys() ]
tuplas_ordenadas = sorted(tuplas, reverse=True) # reverse=True pede uma ordenação em ordem decrescente
palavras = [ t[1] for t in tuplas_ordenadas ]
contagens = [ t[0] for t in tuplas_ordenadas ]

# print(len([c for c in contagens if c < 50]))




n_palavras = 60
eixo_x = np.arange(n_palavras)
plt.figure(figsize=(14,3))
plt.bar(eixo_x[0:n_palavras], contagens[0:n_palavras])
plt.xticks(eixo_x[0:n_palavras], palavras[0:n_palavras], rotation=90)
plt.ylabel('Contagem das palavras')
plt.xlabel('Words')
plt.title('Quantas vezes a palavra aparece na coleção de documentos')
plt.show()


#PLOTANDO GRAFICO 2

#Passo 1: executar vectorizer ao longo de toda a coleção
reviews = [s for s in df['review']]
vectorizer = CountVectorizer(stop_words = sw, binary=True)
x = vectorizer.fit_transform(reviews)
x_ = np.mean(x, axis=0)

# Passo 2: escolher somente elementos correspondentes a cada categoria
filtro_pos = df['review_score']== 1
x_pos = x[filtro_pos,:] # Escolho linhas com sentimento positivo
x_pos_ = np.mean(x_pos, axis=0)
filtro_neg = df['review_score']== 0
x_neg = x[filtro_neg,:] # Escolho linhas com sentimento negativo
x_neg_ = np.mean(x_neg, axis=0)

# Passo 3: ordenar palavras de acordo com a soma geral. Vou incluir tambem a contagem nos grupos positivo e negativo!
tuplas = [ (\
            x_[0,vectorizer.vocabulary_[i]],\
            i,\
            x_pos_[0, vectorizer.vocabulary_[i]],\
            x_neg_[0, vectorizer.vocabulary_[i]]\
            ) for i in vectorizer.vocabulary_.keys() ]
tuplas_ordenadas = sorted(tuplas, reverse=True)
palavras = [ t[1] for t in tuplas_ordenadas ]
contagens_pos = [ t[2] for t in tuplas_ordenadas ]
contagens_neg = [ t[3] for t in tuplas_ordenadas ]

eixo_x = np.arange(len(palavras))
n_palavras = 30

plt.figure(figsize=(14,3))
plt.bar(eixo_x[0:n_palavras]-0.125, contagens_pos[0:n_palavras], width=0.25, label='positive')
plt.bar(eixo_x[0:n_palavras]+0.125, contagens_neg[0:n_palavras],  width=0.25, label='negative')
plt.legend()
plt.xticks(eixo_x[0:n_palavras], palavras[0:n_palavras], rotation=90)
plt.ylabel('P(word | class)')
plt.xlabel('Words')
plt.title('Probability of finding a word in a random document from known class')
plt.show()

````

#### Como podemos observar no primeiro grafico, as palavras que mais aparecem são: produto, prazo e entrega. Dado esse fato podemos concluir que os aspectos mais importantes sob a perpectiva do cliente são o produto em sí, e a entrega. Podemos analisar a palavra atendimento aparece muito no fim do gráfico, mostrando que não é um fator decisivo para o consumidor. <br> 

Agora vamos analisar apenas os reviews negativos, para tentar descobrir aspectos frequentemente mencionados como algo negativo

````python
negativo = df.loc[df["review_score"] == 0]
avaliacões = [av for av in negativo["review"]]
vectorizer = CountVectorizer(stop_words = sw, ngram_range = (1,4))
y = vectorizer.fit_transform(avaliacões)
y_ = np.sum(y, axis=0)

tuplas = [ (y_[0,vectorizer.vocabulary_[i]], i) for i in vectorizer.vocabulary_.keys() ]
tuplas_ordenadas = sorted(tuplas, reverse=True) # reverse=True pede uma ordenação em ordem decrescente
palavras = [ t[1] for t in tuplas_ordenadas ]
contagens = [ t[0] for t in tuplas_ordenadas ]

# print(len([c for c in contagens if c < 50]))


n_palavras = 60
eixo_y = np.arange(n_palavras)
plt.figure(figsize=(5,14))
plt.barh(eixo_y[0:n_palavras], contagens[0:n_palavras])
plt.title("Unigramas que mais aparecem nos reviews negativos")
plt.yticks(eixo_y[0:n_palavras], palavras[0:n_palavras], rotation=0)
plt.show()
````

A palavra produto é a mais mencionada, porém como todas as avaliaçoes se tratam de um produto podemos descosidera-la. Existem várias palavras que fazem menção a entrega, como prazo, entrega, veio, dia, frete. Então podemos perceber que é um fator relevante para uma compra avaliada negativamente

### Ao analisar os bigramas e trigramas, podemos perceber que a entrega/não recebimento tem uma grande quantidade de menções. Dessa forma, vamos investigar mais a entrega dos produtos, tentando entender quais são os fatores relacionados a entrega que os clientes não gostam

## Analise quantitiva relacionada à entrega

Abrindo os df relacionados aos pedidos e aos reviews desses pedidos

````python
df_reviews = pd.read_csv("olist_order_reviews_dataset.csv")
df_orders = pd.read_csv("olist_orders_dataset.csv")
````

Dividindo o df de reviews entre positivos e negativos, considerando o limiar de 3 estrelas.

````python
df_negativos = df_reviews.loc[df_reviews["review_score"] <= 3]
df_positivos = df_reviews.loc[df_reviews["review_score"] > 3]
````

## Hipótese 1: Compras que demoram a chegar tendem a ter avaliações ruins.

Para fazer isso precisaremos calcular o tempo de entrega, usando o dataset de ordens e adicionar essa nova informação no df com as avaliações. <br>
```Tempo de Entrega = Data chegada - data compra```

Mudando as colunas com datas para o formato certo

````python
df_orders["order_purchase_timestamp"] =  pd.to_datetime(df_orders["order_purchase_timestamp"]).dt.date
df_orders["order_approved_at"] =  pd.to_datetime(df_orders["order_approved_at"]).dt.date
df_orders["order_delivered_carrier_date"] =  pd.to_datetime(df_orders["order_delivered_carrier_date"]).dt.date
df_orders["order_delivered_customer_date"] =  pd.to_datetime(df_orders["order_delivered_customer_date"]).dt.date
df_orders["order_estimated_delivery_date"] =  pd.to_datetime(df_orders["order_estimated_delivery_date"]).dt.date
````

Criando uma nova coluna com a informação sobre o tempo de entrega

````python
df_orders["Tempo Entrega"] = df_orders["order_delivered_customer_date"] - df_orders["order_purchase_timestamp"]
````

````python
df_orders["Tempo Entrega int"] = pd.to_numeric(df_orders["Tempo Entrega"].dt.days, downcast='integer')
````

Agora precisamos colocar o tempo de entrega no df de reviews, para isso vamos usar o order_id

````python
df_reviews_tempo = df_reviews.merge(df_orders, left_on = "order_id",  right_on="order_id")
````

````python
df_reviews_tempo = df_reviews_tempo[["review_score","Tempo Entrega" , "Tempo Entrega int" ]]
````

````python
df_reviewTempo_negativos = df_reviews_tempo.loc[df_reviews_tempo["review_score"] <= 3]
df_reviewTempo_positivos = df_reviews_tempo.loc[df_reviews_tempo["review_score"] > 3]
````

Dividindo o df de reviews entre positivos e negativos, considerando o limiar de 3 estrelas.

````python
df_reviewTempo_negativos = df_reviewTempo_negativos.dropna()
df_reviewTempo_positivos = df_reviewTempo_positivos.dropna()
````

````python
media_negativos = df_reviewTempo_negativos["Tempo Entrega int"].mean()
media_positivos = df_reviewTempo_positivos["Tempo Entrega int"].mean()
print("A media de entrega dos produtos mal avaliados foram de {}, e a media de entrega dos bem avaliados foi de {}".format(media_negativos,media_positivos))
````

````python
data = [df_reviewTempo_negativos["Tempo Entrega int"], df_reviewTempo_positivos["Tempo Entrega int"]]
fig, ax = plt.subplots()
ax.boxplot(data)
ax.set_xticklabels(["Negativos", "positivos"])
plt.show()
````

### Como podemos observar, os produtos que não foram bem avaliados tem um tempo de entrega muito maior que os bem avaliados, quase 50% maiores

## Hipótese 2: Compras na qual o tempo de entrega foi subestimado, tendem a ter avaliações ruins.

De forma analaga ao teste de hipótese anterior, vamos avaliar a diferença entre o prazo estimado e a data real da entrega

```Delta tempo = Data entrega - Data Estimada```

````python
df_orders["Tempo_real_Tempo_estimado"] = df_orders["order_delivered_customer_date"] - df_orders["order_estimated_delivery_date"]
````

juntado os df

````python
df_reviews_tempo_Estimado = df_reviews.merge(df_orders, left_on = "order_id",  right_on="order_id")
````

````python
df_reviews_tempo_Estimado["Tempo_real_Tempo_estimado"] = pd.to_numeric(df_reviews_tempo_Estimado["Tempo_real_Tempo_estimado"].dt.days, downcast='integer')
````

Selecionando as ordens que atrasaram

````python
df_atrasado = df_reviews_tempo_Estimado.loc[df_reviews_tempo_Estimado["Tempo_real_Tempo_estimado"] >= 0]
````

````python
df_atrasados_negativos = df_atrasado.loc[df_atrasado["review_score"] <= 3]
````

````python
df_atrasados_positivos = df_atrasado.loc[df_atrasado["review_score"] > 3]
````

````python
count_negativos = len(df_atrasados_negativos)
count_positivos = len(df_atrasados_positivos)
tipo = ["positivo", "negativo"]
quantidade = [count_positivos,count_negativos]
plt.pie(quantidade,labels = tipo, autopct = '%1.1f%%')
plt.title("Avaliacões entre as entregas atrasadas")
plt.legend()
plt.show()
````

````python
print("Ordens atrasadas representam {} % do total de entregas"\
      .format(((count_negativos + count_positivos)/len(df_orders["Tempo_real_Tempo_estimado"]))*100))
````

Como podemos analisar no gráfico, a grande maioria dos pedidos atrasados tem uma avaliação ruim.

````python
media_atraso_negativos = df_atrasados_negativos["Tempo_real_Tempo_estimado"].mean()
media_atraso_positivos = df_atrasados_positivos["Tempo_real_Tempo_estimado"].mean()
````

````python
df_atrasado2 = df_reviews_tempo_Estimado.loc[df_reviews_tempo_Estimado["Tempo_real_Tempo_estimado"] >= 7]
df_atrasados2_negativos = df_atrasado2.loc[df_atrasado2["review_score"] <= 3]
df_atrasados2_positivos = df_atrasado2.loc[df_atrasado2["review_score"] > 3]
````

````python
lista_total = []
lista_neg = []
lista_pos = []
for e in range(0,21):
    df_atrasado2 = df_reviews_tempo_Estimado.loc[df_reviews_tempo_Estimado["Tempo_real_Tempo_estimado"] >= e]
    df_atrasados2_negativos = df_atrasado2.loc[df_atrasado2["review_score"] <= 3]
    df_atrasados2_positivos = df_atrasado2.loc[df_atrasado2["review_score"] > 3]
    count_neg = len(df_atrasados2_negativos)
    count_pos = len(df_atrasados2_positivos)
    total = count_pos + count_neg
    lista_total.append(total)
    lista_neg.append(count_neg/total)
    lista_pos.append(count_pos/total)
````

````python
width = 0.35
fig = plt.figure()
ind = np.arange(21) # the x locations for the groups
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, lista_neg, width, color='r')
ax.bar(ind, lista_pos, width,bottom=lista_neg, color='b')
ax.set_ylabel('proporção de reviews')
ax.set_title('Atraso x reviews')
ax.set_xticks(ind)
ax.set_xlabel('Dias de atraso')
ax.set_yticks(np.arange(0, 1.1, 0.1))
ax.legend(labels=['Reviews Negativos', 'Reviews Positivos'])
plt.show()
````

 ###### Como podemos observar, apenas 7,75% das entregas chegam atrasadas. Apesar dessa número ser baixo, analisando somente as entregas atrasadas, é possível perceber que as avaliações são majoritariamentes negativas. Além disso, nota-se que quanto maior o atraso maior é taxa de avaliações negativas. Isso é valido para atrasos de até uma semana, após esse período, as avaliações negativas se estabilizam em uma faixa de aproximadamente 85%

## Conclusão

####  O tempo de entrega se mostrou um fator relevante para a avaliação do produto. Os itens bem avaliados tem tempo de entrega  quase 50% menores do que os mal avaliados. Assim, a empresa poderia fazer um maior investimento em seu departamento logístico, diminuindo o tempo de entrega e por consequência um menor número de avaliações ruins. <br>Outro fator analisado foi o quanto o atraso da entrega afeta em uma avaliação. De forma geral, as previsões dos marketplace, tendem a superestimar a entrega, de modo que em média as entregas chegam muito antes do esperado. <br>Apesar disso, fazendo uma análise mais profunda dos dados, podemos ver que quando o marketplace subestima o prazo as avaliações tendem a ser ruins, com o número de críticas aumentando conforme o prazo aumenta. Partindo dessa premissa, o marketplace deve buscar nunca subestimar o prazo, pois gera forte descontentamento ao cliente.
